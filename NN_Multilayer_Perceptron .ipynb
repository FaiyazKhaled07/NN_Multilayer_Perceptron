```python
import pandas as pd

data=pd.read_csv('dataset.csv')
data.head(5)
```

```python
#name of all the columns 

print(data.columns)
```

```python
#printing the number of uniques values of all the columns

print('Number of unique values of ID:', data['ID'].nunique())
print('Number of unique values of SurfaceR:', data['SurfaceR'].nunique())
print('Number of unique values of NumberR:', data['NumberR'].nunique())
print('Number of unique values of TypeR:', data['TypeR'].nunique())
print('Number of unique values of VegetationR:', data['VegetationR'].nunique())
print('Number of unique values of Surroundings1:', data['Surroundings1'].nunique())
print('Number of unique values of Surroundings2:', data['Surroundings2'].nunique())
print('Number of unique values of Surroundings3:', data['Surroundings3'].nunique())
print('Number of unique values of UseR:', data['UseR'].nunique())
print('Number of unique values of FishingR:', data['FishingR'].nunique())
print('Number of unique values of AcessR:', data['AcessR'].nunique())
print('Number of unique values of RoadDistanceR:', data['RoadDistanceR'].nunique())
print('Number of unique values of BuildingR:', data['BuildingR'].nunique())
print('Number of unique values of PollutionR:', data['PollutionR'].nunique())
print('Number of unique values of ShoreR:', data['ShoreR'].nunique())
print('Number of unique values of Green frogs:', data['Green frogs'].nunique())
print('Number of unique values of Brown frogs:', data['Brown frogs'].nunique())
print('Number of unique values of Common toad:', data['Common toad'].nunique())
print('Number of unique values of Fire-bellied toad:', data['Fire-bellied toad'].nunique())
print('Number of unique values of Tree frog:', data['Tree frog'].nunique())
print('Number of unique values of Common newt:', data['Common newt'].nunique())
print('Number of unique values of Great crested newt:', data['Great crested newt'].nunique())
```

```python
#dropping multiple columns

new_data=data.drop(columns=['ID', 'Surroundings1','Surroundings2','Surroundings3','Common toad','Fire-bellied toad','Tree frog','Common newt','Great crested newt'], axis=1)
new_data.head(5)
```

```python
#converting the two categorical features to numerical features 

final_data=new_data.copy()
final_data=pd.get_dummies(final_data, columns=['TypeR','ShoreR'])
print(final_data.head(2))
```

Now from all the habitats I will be choosing five most important habitats as my input features and 'Green frogs', 'Brown frogs' will be my target features.


```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier()

#for green frogs
x1=final_data[['SurfaceR','NumberR','VegetationR','UseR','FishingR','AcessR','RoadDistanceR','BuildingR','PollutionR','TypeR_a','TypeR_b','TypeR_d','TypeR_e','TypeR_g','TypeR_h','TypeR_i','TypeR_j','ShoreR_concrete','ShoreR_natural']].values
y1=final_data['Green frogs']

#feature importance for green frogs
print("Feature importance for Green frogs: \n ")

dt1=DecisionTreeClassifier(random_state=1, criterion='entropy', max_depth=10)
dt1.fit(x1,y1)

for i,column in enumerate(final_data[['SurfaceR','NumberR','VegetationR','UseR','FishingR','AcessR','RoadDistanceR','BuildingR','PollutionR','TypeR_a','TypeR_b','TypeR_d','TypeR_e','TypeR_g','TypeR_h','TypeR_i','TypeR_j','ShoreR_concrete','ShoreR_natural']]): 
    print(('feature importance for {} is : {}'.format(column, dt1.feature_importances_[i])))
    
    
#for brown frogs    
x2=final_data[['SurfaceR','NumberR','VegetationR','UseR','FishingR','AcessR','RoadDistanceR','BuildingR','PollutionR','TypeR_a','TypeR_b','TypeR_d','TypeR_e','TypeR_g','TypeR_h','TypeR_i','TypeR_j','ShoreR_concrete','ShoreR_natural']].values
y2=final_data['Brown frogs']

#feature importance for brown frogs
print("Feature importance for brown frogs: \n ")

dt2=DecisionTreeClassifier(random_state=1, criterion='entropy', max_depth=10)
dt2.fit(x2,y2)

for i,column in enumerate(final_data[['SurfaceR','NumberR','VegetationR','UseR','FishingR','AcessR','RoadDistanceR','BuildingR','PollutionR','TypeR_a','TypeR_b','TypeR_d','TypeR_e','TypeR_g','TypeR_h','TypeR_i','TypeR_j','ShoreR_concrete','ShoreR_natural']]): 
    print(('feature importance for {} is : {}'.format(column, dt2.feature_importances_[i])))
```

```python
from sklearn.model_selection import train_test_split

#for green frogs
descriptive_features1=final_data[['SurfaceR', 'VegetationR', 'RoadDistanceR', 'FishingR', 'UseR']]
target_feature1=final_data['Green frogs']

#train-test split
X1_train, X1_test, y1_train, y1_test=train_test_split(descriptive_features1, target_feature1, test_size=0.2, random_state=1)

#for brown frogs
descriptive_features2=final_data[['SurfaceR', 'RoadDistanceR', 'FishingR', 'TypeR_h', 'VegetationR']]
target_feature2=final_data['Brown frogs']

#train-test split
X2_train, X2_test, y2_train, y2_test=train_test_split(descriptive_features2, target_feature2, test_size=0.2, random_state=1)
```

Using linear activation function for both target features then training the model and finding accuracy.

```python
#linear activation 

from sklearn.neural_network import MLPClassifier
import sklearn.metrics as metric

#for green frogs
#using Multi-layer Perceptron classifier(MLP) to create net.
linear1 = MLPClassifier(activation='identity')
linear1.fit(X1_train, y1_train)

prediction1=linear1.predict(X1_test)
print('Accuracy for green frogs:', linear1.score(X1_test,y1_test))

#for brown frogs
#using Multi-layer Perceptron classifier(MLP) to create net.
linear2 = MLPClassifier(activation='identity')
linear2.fit(X2_train, y2_train)

prediction2=linear2.predict(X2_test)
print('Accuracy for brown frogs:', linear2.score(X2_test,y2_test))
```

Using non-linear activation function for both target features then training the model and finding accuracy.

```python
#non-linear activation 

from sklearn.neural_network import MLPClassifier
import sklearn.metrics as metric

#for green frogs
#using Multi-layer Perceptron classifier(MLP) to create net.
nonlinear1 = MLPClassifier(activation='logistic')
nonlinear1.fit(X1_train, y1_train)

prediction1=nonlinear1.predict(X1_test)
print('Accuracy for green frogs:', nonlinear1.score(X1_test,y1_test))

#for brown frogs
#using Multi-layer Perceptron classifier(MLP) to create net.
nonlinear2 = MLPClassifier(activation='logistic')
nonlinear2.fit(X2_train, y2_train)

prediction2=nonlinear2.predict(X2_test)
print('Accuracy for brown frogs:', nonlinear2.score(X2_test,y2_test))
```

Now training three more Neural Networks with different parameters and finding accuracy.

```python
#for green frogs
#using Multi-layer Perceptron classifier(MLP) to create net.
mynet1 = MLPClassifier(hidden_layer_sizes=200, activation='relu', solver='adam', max_iter=500)
mynet1.fit(X1_train, y1_train)

prediction1=mynet1.predict(X1_test)
print('Accuracy for green frogs:', mynet1.score(X1_test,y1_test))

#for brown frogs
#using Multi-layer Perceptron classifier(MLP) to create net.
mynet2 = MLPClassifier(hidden_layer_sizes=200, activation='relu', solver='adam', max_iter=500)
mynet2.fit(X2_train, y2_train)

prediction2=mynet2.predict(X2_test)
print('Accuracy for brown frogs:', mynet2.score(X2_test,y2_test))
```
